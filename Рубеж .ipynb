Имеем запаздывание

````python
diff_arr = []
for i in range (5):
  diff_arr.append(fpg_arr[i] - t_arr[i])
print(diff_arr)
````

Среднее значение опаздывания

````python
sum = 0
for i in diff_arr:
  sum += i

average = sum/5
print(average)
````
